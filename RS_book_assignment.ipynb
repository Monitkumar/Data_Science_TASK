## Data Preparation

```python
book_recomm = pd.pivot_table(data = new_book_data,values='BookRating',index='UserId',columns='BookTitle').fillna(0)
book_recomm
```

```python
book_recomm.head()
```

```python
book_recomm = book_recomm.reset_index(drop = True)
book_recomm
```

```python
book_recomm.index = new_book_data['UserId'].unique()
```

```python
book_recomm
```

## Model Building and Finding out the similar user for recommendations

### 1. Using Correlation Matrix for UBCF

```python
book_transpose = book_recomm.T
book_transpose
```

```python
book_recomm_corr = book_transpose.corr().round(2)
```

```python
book_recomm_corr
```

## Using cosine similarty score

```python
from sklearn.metrics import pairwise_distances
```

```python
book_recomm
```

```python
book_recomm.values
```

```python
reader_to_reader = pairwise_distances(X = book_recomm.values , metric='euclidean')
reader_to_reader
```

```python
book_recomm.index
```

```python
reader_df = pd.DataFrame(data=reader_to_reader,columns=book_recomm.index,index = book_recomm.index)
reader_df
```

```python
reader_cosine = 1 - pairwise_distances(X = book_recomm.values,metric ='cosine')
reader_cosine
```

```python
reader_cf = pd.DataFrame(data=reader_cosine,columns=book_recomm.index,index = book_recomm.index)
reader_cf
```

```python
np.fill_diagonal(reader_cf.to_numpy(), 0)
```

```python
reader_cf
```

```python
reader_cf.idxmax()
```

```python
reader_top30 = reader_cf.iloc[:30,:30]
reader_top30
```
