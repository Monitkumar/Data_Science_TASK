```python
# Applying one hot encoding
bank_data2=pd.get_dummies(bank_data,columns=['job','marital','education','contact','poutcome','month'])
bank_data2
```

```python
bank_data2.info()
```

```python
bank_data2.default.map(dict(yes=1, no=0))
bank_data2.housing.map(dict(yes=1, no=0))
bank_data2.loan.map(dict(yes=1, no=0))
bank_data2.y.map(dict(yes=1, no=0))


bank_data2
```

## Model building

```python
X = bank_data2.drop(labels = 'y',axis =1)
y = bank_data2[['y']]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

## Model Training

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train,y_train)
```

## Model Testing

```python
y_pred_train = logistic_model.predict(X_train)
y_pred_train
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_train,y_pred_train)
```

```python
confusion_matrix(y_train,y_pred_train)
```

## Test data

```python
y_test_pred = logistic_model.predict(X_test)
y_test_pred
```

```python
accuracy_score(y_test,y_test_pred)
```

```python
confusion_matrix(y_test,y_test_pred)
```

```python
print(classification_report(y_test,y_test_pred))
```
