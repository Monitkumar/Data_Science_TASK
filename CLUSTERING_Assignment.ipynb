```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
cr_data =  pd.read_csv('crime_data.csv')
cr_data
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
cr_hierarchical = norm_func(cr_data.iloc[:,1:])
```

```python
cr_hierarchical
```

```python
dendrogram = sch.dendrogram(sch.linkage(cr_hierarchical, method='single'))
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'single')
```

```python
y_hc = hc.fit_predict(cr_hierarchical)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
y_hc
```

```python
cr_hierarchical['clusterid_hierarchical'] = y_hc
cr_hierarchical.head()
```

## Using K means Clustering

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
cr_data_kmeans = pd.read_csv("crime_data.csv")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_cr_kmeans = scaler.fit_transform(cr_data_kmeans.iloc[:,1:])
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_cr_kmeans)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
clusters_kmeans = KMeans(4, random_state=42)
clusters_kmeans.fit(scaled_cr_kmeans)
```

```python
clusters_kmeans.labels_
```

```python
cr_data_kmeans['clusterid_Kmeans'] = clusters_kmeans.labels_
```

```python
clusters_kmeans.cluster_centers_
```

```python
cr_data_kmeans.groupby('clusterid_Kmeans').agg(['mean']).reset_index()
```

```python
cr_data_kmeans
```
