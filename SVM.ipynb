```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,roc_auc_score
```

```python
import warnings
warnings.filterwarnings('ignore')
```

## Forestfires data

```python
fire_data = pd.read_csv('forestfires.csv')
fire_data.head()
```

```python
fire_data.shape
```

```python
fire_data.describe()
```

```python
fire_data.isna().sum()
```

```python
fire_data.dtypes
```

```python
fire_data.drop(["month","day"],axis =1 ,inplace =True)
```

```python
fire_data
```

```python
le = LabelEncoder()
fire_data['size_category'] = le.fit_transform(fire_data.size_category)
```

```python
fire_data.head()
```

```python
X = fire_data.drop('size_category',axis=1)
y = fire_data[['size_category']]
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=15,stratify = y)
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
x_scaled
```

```python
X_train_scaled,X_test_scaled,y_train_scaled,y_test_scaled = train_test_split(x_scaled,y,test_size=0.20,random_state=15,stratify = y)
```

## kernel = linear

```python
fire_svc_scaled = SVC(kernel='linear')
fire_svc_scaled.fit(X_train_scaled,y_train_scaled)
y_pred_scaled = fire_svc_scaled.predict(X_test_scaled)
```

```python
print("Overall Accuracy : ",round(accuracy_score(y_test,y_pred_scaled),4))
print("Precision        : ",round(precision_score(y_test,y_pred_scaled),4))
print("Recall           : ",round(recall_score(y_test,y_pred_scaled),4))
print("AUC Score        : ",round(roc_auc_score(y_test,y_pred_scaled),4))
print("Confusion Matrix :\n",confusion_matrix(y_test,y_pred_scaled))
```

## kernel = rbf

```python
fire_svc_scaled = SVC(kernel='rbf')
fire_svc_scaled.fit(X_train_scaled,y_train_scaled)
y_pred_scaled = fire_svc_scaled.predict(X_test_scaled)
```

```python
print("Overall Accuracy : ",round(accuracy_score(y_test,y_pred_scaled),4))
print("Precision        : ",round(precision_score(y_test,y_pred_scaled),4))
print("Recall           : ",round(recall_score(y_test,y_pred_scaled),4))
print("AUC Score        : ",round(roc_auc_score(y_test,y_pred_scaled),4))
print("Confusion Matrix :\n",confusion_matrix(y_test,y_pred_scaled))
```

## kernel = polynomial

```python
fire_svc_scaled = SVC(kernel='poly')
fire_svc_scaled.fit(X_train_scaled,y_train_scaled)
y_pred_scaled = fire_svc_scaled.predict(X_test_scaled)
```

```python
print("Overall Accuracy : ",round(accuracy_score(y_test,y_pred_scaled),4))
print("Precision        : ",round(precision_score(y_test,y_pred_scaled),4))
print("Recall           : ",round(recall_score(y_test,y_pred_scaled),4))
print("AUC Score        : ",round(roc_auc_score(y_test,y_pred_scaled),4))
print("Confusion Matrix :\n",confusion_matrix(y_test,y_pred_scaled))
```

# Salary Dataset

```python
Salary_Train = pd.read_csv("SalaryData_Train(1).csv")
Salary_Train
```

```python
Salary_Test = pd.read_csv("SalaryData_Test(1).csv")
Salary_Test
```

```python
Salary_data = pd.concat([Salary_Test,Salary_Train])
Salary_data
```

```python
Salary_data.dtypes
```

```python
Salary_data.isnull().sum()
```

```python
del Salary_data['educationno']
```

```python
Salary_data.shape
```

```python
le = LabelEncoder()
Salary_data['workclass'] = le.fit_transform(Salary_data.workclass)
Salary_data['education'] = le.fit_transform(Salary_data.education)
Salary_data['maritalstatus'] = le.fit_transform(Salary_data.maritalstatus)
Salary_data['relationship'] = le.fit_transform(Salary_data.relationship)
Salary_data['sex'] = le.fit_transform(Salary_data.sex)
Salary_data['race'] = le.fit_transform(Salary_data.race)
Salary_data['native'] = le.fit_transform(Salary_data.native)
Salary_data['Salary'] = le.fit_transform(Salary_data.Salary)
Salary_data['occupation'] = le.fit_transform(Salary_data.occupation)
```

```python
Salary_data
```

```python
X = Salary_data.drop('Salary',axis=1)
y = Salary_data['Salary']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=15,stratify = y)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
x_scaled
```

```python
X_train_scaled,X_test_scaled,y_train_scaled,y_test_scaled = train_test_split(x_scaled,y,test_size=0.20,random_state=15,stratify = y)
X_train_scaled.shape,X_test_scaled.shape,y_train_scaled.shape,y_test_scaled.shape
```

```python
svc_classifier = SVC(kernel='rbf')
svc_classifier.fit(X_train_scaled,y_train_scaled)
y_pred_scaled = svc_classifier.predict(X_test_scaled)
```

```python
print("Overall Accuracy : ",round(accuracy_score(y_test,y_pred_scaled),4))
print("Precision        : ",round(precision_score(y_test,y_pred_scaled),4))
print("Recall           : ",round(recall_score(y_test,y_pred_scaled),4))
print("AUC Score        : ",round(roc_auc_score(y_test,y_pred_scaled),4))
print('Confusion Matrix :\n',confusion_matrix(y_test,y_pred_scaled))
```

```python
svc_classifier = SVC(kernel='poly')
svc_classifier.fit(X_train_scaled,y_train_scaled)
y_pred_scaled = svc_classifier.predict(X_test_scaled)
```

```python
print("Overall Accuracy : ",round(accuracy_score(y_test,y_pred_scaled),4))
print("Precision        : ",round(precision_score(y_test,y_pred_scaled),4))
print("Recall           : ",round(recall_score(y_test,y_pred_scaled),4))
print("AUC Score        : ",round(roc_auc_score(y_test,y_pred_scaled),4))
print('Confusion Matrix :\n',confusion_matrix(y_test,y_pred_scaled))
```
