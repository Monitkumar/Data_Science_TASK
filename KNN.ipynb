```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,r2_score
```

## Using glass dataset

```python
glass_data = pd.read_csv('glass.csv')
glass_data
```

```python
glass_data.shape
```

```python
glass_data.describe()
```

```python
glass_data.isna().sum()
```

```python
X = glass_data.drop(['Type'],axis=1)
y = glass_data['Type']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state= 15,stratify=y)
```

```python
glass_knn_model = KNeighborsClassifier()
glass_knn_model.fit(X_train,y_train)
```

```python
y_pred = glass_knn_model.predict(X_test)
y_pred
```

```python
print("Accuracy score   :",round(accuracy_score(y_test,y_pred),4))
print("R2 Score         :",round(r2_score(y_test,y_pred),4))
```

```python
import warnings
warnings.filterwarnings('ignore')
neighbours = list(range(1,50,2))
cv_scores = []

for i in neighbours:
    Glass_knn_model = KNeighborsClassifier(n_neighbors = i)
    cv_score = cross_val_score(estimator = glass_knn_model,X=X,y=y,cv=10)
    cv_scores.append(cv_score.mean())
```

```python
cv_scores
```

```python
neighbours[cv_scores.index(max(cv_scores))]
```

## As per above line to get better result K should be K=1

```python
glass_knn_model = KNeighborsClassifier(n_neighbors=1)
glass_knn_model.fit(X_train,y_train)
y_pred = glass_knn_model.predict(X_test)
print("Accuracy score   :", round(accuracy_score(y_test,y_pred),4))
print("R2 Score         :",round(r2_score(y_test,y_pred),4))
```

##  If we take K=5 then Accuracy is 0.6279
## If we take K=1 then Accuracy is 0.6744

# Using zoo data

```python
zoo_data = pd.read_csv('Zoo.csv')
zoo_data
```

```python
zoo_data.describe()
```

```python
zoo_data.isnull().sum()
```

```python
zoo_Label = LabelEncoder()
zoo_data['animal name'] = zoo_Label.fit_transform(zoo_data['animal name'])
```

```python
zoo_data
```

```python
X = zoo_data.drop(['type'],axis=1)
y = zoo_data['type']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state= 15,stratify=y)
```

```python
zoo_knn_model = KNeighborsClassifier()
zoo_knn_model.fit(X_train,y_train)
```

```python
y_pred = zoo_knn_model.predict(X_test)
y_pred
```

```python
print("Accuracy score   : ", round(accuracy_score(y_test,y_pred),4))
print("R2 Score         :",round(r2_score(y_test,y_pred),4))
```

```python
neighbours = list(range(1,50,2))
cv_scores = []

for i in neighbours:
    zoo_knn_model = KNeighborsClassifier(n_neighbors = i)
    cv_score = cross_val_score(estimator = zoo_knn_model,X=X,y=y,cv=10)
    cv_scores.append(cv_score.mean())
```

```python
cv_scores
```

```python
neighbours[cv_scores.index(max(cv_scores))]
```

```python
zoo_knn_model = KNeighborsClassifier(n_neighbors=1)
zoo_knn_model.fit(X_train,y_train)
y_pred = zoo_knn_model.predict(X_test)
print("Accuracy score   :", round(accuracy_score(y_test,y_pred),4))
print("R2 Score         :",round(r2_score(y_test,y_pred),4))
```

## If we take K=5 then Accuracy is 0.4286
## If we take K=1 then Accuracy is 0.6667
