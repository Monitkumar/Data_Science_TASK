```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

## Using books dataset

```python
book_data = pd.read_csv('book (2).csv')
book_data
```

## Apriori algorithm

```python
freq_itemsets1 = apriori(book_data, min_support=0.1, use_colnames=True)
freq_itemsets1
```

```python
rule_1 = association_rules(freq_itemsets1, metric="lift", min_threshold=0.7)
rule_1
```

```python
rule_1.sort_values('lift',ascending = False)[0:20]
```

```python
rule_1[rule_1.lift>1]
```

```python
freq_itemsets2 = apriori(book_data, min_support=0.2, use_colnames=True)
freq_itemsets2
```

```python
rule_2 = association_rules(freq_itemsets2, metric="lift", min_threshold=1)
rule_2
```

```python
freq_itemsets3 = apriori(book_data, min_support=0.3, use_colnames=True)
freq_itemsets3
```

```python
rule_3 = association_rules(freq_itemsets3, metric="lift", min_threshold=0.5)
rule_3
```

## Using Movies dataset


```python
import pandas as pd
```

```python
mov_data = pd.read_csv('my_movies (1).csv')
mov_data.shape
```

```python
m_data=pd.get_dummies(mov_data)
m_data.head()
```

```python
freq_itemsets1 = apriori(m_data, min_support=0.5, use_colnames=True)
freq_itemsets1
```

```python
rules1 = association_rules(freq_itemsets1, metric="lift", min_threshold=1)
rules1
```

```python
rules1.sort_values('lift',ascending = False)[0:20]
```

```python
rules1[rules1.lift>1]
```

```python
rules2 = association_rules(freq_itemsets2, metric="lift", min_threshold=1)
rules2
```
