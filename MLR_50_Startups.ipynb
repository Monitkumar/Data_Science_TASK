```python
le = LabelEncoder()
startup_data['state'] = le.fit_transform(startup_data['State'])                                   
```

```python
del startup_data['State']
```

## Model Building

```python
startup_data
```

```python
startup_data2 = StandardScaler()
startup_data2 = startup_data2.fit_transform(startup_data)
```

```python
startup_data2
```

```python
startup_df = pd.DataFrame(data=startup_data2,columns=[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit','state']])
```

```python
startup_df
```

## Model Building

```python
X = startup_df[['R&D Spend', 'Administration', 'Marketing Spend','state']]
y=startup_df[['Profit']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30)
```

## Model training

```python
from sklearn.linear_model import LinearRegression
```

```python
startup_model = LinearRegression()
startup_model.fit(X_train,y_train)
```

## Model testing

```python
y_pred = startup_model.predict(X_test)
y_pred
```

```python
print("R2 Score :",round(r2_score(y_test,y_pred),4))
```

```python
print("MSE      :",round(mean_squared_error(y_test,y_pred),4))
```
